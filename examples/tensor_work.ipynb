```python
from spinbox import *
from spinbox.extras import *

import itertools
```

```python
X1 = SigmaTauCoupling(3)
X1.random(1.,0)
X1.coefficients += 1

sig, tau = sigma_tau_operators_hilbert(3)

def xpi_operator(coupling, i, j):
    out = HilbertOperator(3).zero()
    for a in (0,1,2):
        for b in (0,1,2):
            out += sig[i][a]*sig[j][b].scale(coupling[a,i,b,j])
    return out
```

```python
ii,jj,kk = 0,1,2
new = xpi_operator(X1,ii,jj) * xpi_operator(X1,jj,kk) - xpi_operator(X1,jj,kk) * xpi_operator(X1,ii,jj)
# new = xpi_operator(X1,ii,jj) * xpi_operator(X1,jj,kk)
print(new)
```
